```python
import pandas as pd 
import geopandas as gpd 
df = pd.read_excel('GMaps Data/2025-12-15/supermarkets_in_Milan.xlsx')
```

```python
df
```

```python
# Convert Latitude and Longitude columns to numeric (float), coercing errors to NaN
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

df.dropna(subset=['latitude', 'longitude'], inplace=True)
```

```python
geometry = gpd.points_from_xy(df.longitude, df.latitude)

gdf = gpd.GeoDataFrame(
    df,
    geometry=geometry,
    crs="EPSG:4326"
)
```

```python
m = gdf.explore(
    marker_kwds={
        'radius': 5 , 
        "fillOpacity": 1,
        "weight": 1
    }
)
```

```python
m.save('sample_map.html') #in case you want to save the map as an HTML file
m #display the map in the notebook
```
